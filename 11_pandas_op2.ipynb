# 데이터 선택 & 슬라이싱

**iloc와 loc로 데이터 선택하기**<br>
loc는‘칼럼명’을 기준으로 데이터프레임의 데이터를 선택하고, iloc은 ‘인덱스번호’로 선택한다.

```
import numpy as np
import pandas as pd

df2d = pd.DataFrame(np.random.randn(8, 8), columns=['서울', '경기', '광주', '대구', '부산', '전주', '대전', '세종'])
df2d = np.abs(df2d)
df2d
```

<b>열 추가</b>

```
df2d.loc[:, '인천'] = np.abs(np.random.rand(8))
df2d
```

<b>iloc : 인덱스 번호를 이용하여 특정한 열 추출</b>

```
sudo = df2d.iloc[:, [0,1,8]]
sudo
```

<b>loc : 칼럼명을 이용하여 특정 열 추출하기</b>

```
yeongnam = df2d.loc[:, ['부산', '대구']]
yeongnam
```

```
df2d.loc[:, '평균'] = df2d.iloc[:, :9].mean(axis=1)
df2d
```

<b>조건에 맞는 행 추출하기</b>

```
high_value = df2d.loc[df2d['평균'] > 0.6]
high_value
```

<b>조건에 맞는 행과 열 추출하기</b>

```
high_value_honam = df2d.loc[df2d['평균'] > 0.6, ['광주', '전주']]
high_value_honam
```

<b>loc와 iloc의 차이</b> : 어떤 차이가 또 있을까요?

```
head_loc = df2d.loc[:3]
head_loc
```

```
head_iloc = df2d.iloc[:3]
head_iloc
```

# DataFrame과 Series 사이의 연산

### 2차원 배열과 1차원 배열 사이의 연산

```
import numpy as np
import pandas as pd

arr2d = np.arange(12.).reshape((3, 4))
arr2d
```

```
arr1d = np.arange(4.0)
arr1d
```

```
arr2d - arr1d
```

```
arr2d + arr1d
```

<b>2차원 배열과 그 배열의 한 행의 차를 계산한 결과를 설명해보시오.</b>

### DataFrame과 Series 사이의 연산도 위의 이차원 배열과 1차원 배열 사이의 연산과 비슷하다.


```
df = pd.DataFrame(np.arange(16.).reshape((4, 4)), columns = list('abcd'), index=['서울', '부산', '대구', '광주'])
df
```

```
sr = pd.Series(np.arange(4.), index=list('abcd')) # index 리스트가 dataframe의 columns 리스트와 같아야 한다.
sr
```

연산자(+, -)를 이용한 DataFrame과 Series의 연산

```
df + sr
```

```
df - sr
```

함수(add(), sub())를 이용한 DataFrame과 Series의 연산</br>


```
df.add(sr, axis = 1)
```

```
df.sub(sr, axis = 1)
```
