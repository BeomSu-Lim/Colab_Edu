```
import requests as rq
from bs4 import BeautifulSoup as bs
# 웹 페이지 가져오기
res = rq.get('http://www.andong.ac.kr/main/index.do')
# 가져온 웹 페이지의 데이터 확인
res.content
```

```
# 웹 페이지 파싱하기
soup = bs(res.content, 'html.parser')
# 필요한 데이터 추출하기
mydata = soup.find('div', id='top-banner')
# 추출한 데이터 활용하기
print(mydata.get_text())
```

```
# 필요한 데이터 추출하기
mydata = soup.findAll('p', 'tit')
# 추출한 데이터 활용하기
for string in mydata:
    print(string.get_text())
```
