#전라남도 관광 지도 만들기

###웹 페이지 가져와서 분석하기

```
import requests as rq
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd

url = 'http://apis.data.go.kr/6460000/tourInfo/getTourlnfoList?'
serviceKey = '&ServiceKey=hInmnIlbf0k2%2BJV2mirOFl1mto3bh28WCLHB%2BNkUKdr5s%2FUp91EPM78lplJti92KCqKyTmm1yHjWG4rHLcuHIg%3D%3D'
params = '&pageSize=10&startPage=1&numOfRows=10&tourArea=목포&tourCategory=관광'
openAPI = url+serviceKey+params

res = rq.get(openAPI)
soup = bs(res.content, 'html.parser')
infoAll = soup.findAll('list')
infoAll
```

###원하는 데이터 추출하기

```
name=[] # 관광지명
category2=[] # 구분2
category3=[] # 구분3
addr=[] # 관광지 주소
phone=[] # 전화번호
latitude=[] # 위도
longitude=[] # 경도

for data in infoAll:
    if (data.select_one('tourname')):
        name.append(data.select_one('tourname').text)
    else: 
        name.append('-')

    if (data.select_one('tourcategory2')):
        category2.append(data.select_one('tourcategory2').text)
    else: 
        category2.append('-')

    if (data.select_one('tourcategory3')):
        category3.append(data.select_one('tourcategory3').text)
    else: 
        category3.append('-')

    if (data.select_one('touraddr')):
        addr.append(data.select_one('touraddr').text)
    else: 
        addr.append('-')

    if (data.select_one('tourphone')):
        phone.append(data.select_one('tourphone').text)
    else: 
        phone.append('-')

    if (data.select_one('tourlatitude')):
        latitude.append(data.select_one('tourlatitude').text)
    else: 
        latitude.append('-')

    if (data.select_one('tourlongitude')):
        longitude.append(data.select_one('tourlongitude').text)
    else: 
        longitude.append('-')
```

```
print(len(name), len(category2), len(category3), len(addr), len(phone), len(latitude), len(longitude))
```

###데이터프레임 만들기

```
latitude = np.array(latitude).astype(np.float)
longitude = np.array(longitude).astype(np.float)

dic={'관광지명':name, '분류2' : category2, '분류3':category3, '주소':addr, '전화번호':phone, '위도':latitude, '경도':longitude}
tourJeollanamdo = pd.DataFrame(dic)

tourJeollanamdo
```

###지도 위에 관광지 표시하고 HTML로 저장하기

```
import folium

cityhall = folium.Map(location=[34.81178605219916, 126.39227174698898], zoom_start=13)


for i in range(len(tourJeollanamdo)) :
    location = (latitude[i], longitude[i])
    marker=folium.Marker(location, 
                        tooltip=name[i], 
                        icon=folium.Icon(icon='star', color='red'))
    marker.add_to(cityhall)  # 마커 객체를 지도에 추가함

cityhall    

cityhall.save('/gdrive/My Drive/빅데이터분석전문가과정/data/tourMokpho.html') 
```
