# 객체 지향 프로그래밍
속성(attribute, state)과 행동(behavior, method)을 가지고 있는 독립된 객체들을 선언하고, 그 객체들 사이의 유기적인 상호 작용을 통하여 프로그램이 동작하도록 프로그램을 작성하는 방법

객체(object) : 이 세상 모든 것은 객체이다. 객체는 속성(상태)과 행동(메서드)을 가지고 있다.<br>
**object = attribute + method**

**원 객체**
* 속성 : 중심좌표(x, y), 반지름(radius), 파이(원 넓이와 둘레 계산에 반드시 필요한 자료)
* 메소드 : 둘레 계산, 넓이 계산

**사각형 객체**
* 속성 : 시작 좌표(x1, y1), 끝 좌표(x2, y2)
* 메소드 : 둘레 계산, 넓이 계산

###2. 메서드의 정의

```
class Circle :
    pi = 3.141592
    cx, cy, radius = 0., 0., 0.
    def __init__(self, cx, cy, radius) :
        self.cx = cx
        self.cy = cy
        self.radius = radius

    def calLength(self) :
        return 2 * self.radius * self.pi

    def calArea(self) :
        return self.radius ** 2 * self.pi
```

###3. 클래스 인스턴스의 생성
* 클래스 : 중심좌표와 반지름을 속성으로 가지고 둘레의 길이와 넓이를 계산할 수 있는 원(이미지만 있을뿐 실체가 없다.) 
* 인스턴스(실체) 1 : 중심 좌표가 0,0이고 반지름의 길이가 5인 실제 원
* 인스턴스(실체) 2 : 중심 좌표가 10,10이고 반지름의 길이가 3인 실제 원

```
class Circle :
    pi = 3.141592
    cx, cy, radius = 0., 0., 0.
    def __init__(self, cx, cy, radius) :
        self.cx = cx
        self.cy = cy
        self.radius = radius

    def calLength(self) :
        return 2 * self.radius * self.pi

    def calArea(self) :
        return self.radius ** 2 * self.pi

c1 = Circle(0, 0, 5) # Circle 클래스로부터 c1 인스턴스 생성
c2 = Circle(10, 10, 3)
```

```
# 인스턴스 c1의 둘레 계산
c1.calLength()
```

```
# 인스턴스 c2의 넓이 계산
c2.calArea()
```

## 같이 해보기
사각형 클래스를 선언하고 사각형 인스턴스의 둘레와 넓이를 계산할 수 있는 프로그램을 작성해봅시다.

```
class Ractangle : 
    x1, y1, x2, y2 = 0, 0, 0, 0
    def __init__(self, x1, y1, x2, y2) :
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2 

    def calPerimeter(self): 
        return 2*(abs(self.x2 - self.x1) + abs(self.y2 - self.y1))

    def calArea(self):
        return abs(self.x2 - self.x1) * abs(self.y2 - self.y1)

r1 = Ractangle(150, 180, 100, 100)
print("원1의 둘레 =",r1.calPerimeter())
print("원1의 넓이 =",r1.calArea())

r2 = Ractangle(0, 0, 280, 240)
print("원2의 둘레 =",r2.calPerimeter())
print("원2의 넓이 =",r2.calArea())
```

```
import matplotlib.pyplot as plt

center = (10, 10) # 중심 좌표
radius = 5.0 # 반지름

# 중심 좌표 (10,10), 반지름(5), 채우기 색(노란색), 테두리 색(빨간색)을 가진 원 객체(인스턴스) 생성
c = plt.Circle(center, radius, fc='y', ec='r')

# 좌표 평면 생성
ax = plt.axes(xlim=(0, 20), ylim=(0, 20))
ax.add_patch(c)             # 원 객체 c를 패치
ax.set_aspect('equal')      # X축과 Y축의 비율을 1:1로 설정

plt.show()                  # 좌표 평면에 원 그리기
```

###4. 상속(inheritance)

**상속** 
* 부모 클래스(물려주는 클래스)와 자식 클래스(물려받는 객체)가 있다.
* 자식 클래스를 선언할 때 괄호 안에 부모 클래스를 명시한다.
* 자식 클래스는 부모 클래스의 속성과 메서드를 모두 물려 받는다.


>class 부모 :
>>...속성과 메서드 ...


>class 자식(부모) :
>>...속성과 메서드 ...

```
class Person :
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def get_info(self):
        return self.name, self.age, self.gender

class Student(Person) :
    def __init__(self, name, age, gender, score):
        self.name = name
        self.age = age
        self.gender = gender
        self.score = score

    def get_result(self):
        if self.score >= 70 :
            return "합격"
        else :
            return "불합격"

st1 = Student('홍길동', 16, '남자', 85)
name, age, gender = st1.get_info()
result = st1.get_result()
print(str(age)+"세 "+ gender+"인 "+name+"씨 "+result+"입니다.")

st1 = Student('박인영', 16, '여자', 68)
name, age, gender = st1.get_info()
result = st1.get_result()
print(str(age)+"세 "+ gender+"인 "+name+"씨 "+result+"입니다.")
```
