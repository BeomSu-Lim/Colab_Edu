### (1) 엑셀 파일 열기

```
import pandas as pd
cctvSeoul = pd.read_excel('/gdrive/My Drive/빅데이터교육/seoul_cctv.xlsx')
cctvSeoul.head(3)
```

####'기관명' 칼럼의 데이터에서 공백 제거

```
guName = cctvSeoul['기관명']
newName=[]
for name in guName:
    newName.append(name.replace(' ', ''))
newName
```

#### '기관명' 칼럼을 삭제하고, 새로 만들어둔 newName을 dataframe의 인덱스로 설정

```
del cctvSeoul['기관명']
cctvSeoul.index = newName
cctvSeoul.head()
```

#### '2011년 이전' 칼럼부터 '2014년' 칼럼까지의 합계를 구한 후 ''2014년 이전'이라는 칼럼 이름으로 삽입

```
cctvSeoul.loc[:, '2014년 이전'] = cctvSeoul.iloc[:, 1:5].sum(axis=1)
cctvSeoul.head()
```

#### '2011년 이전' 칼럼부터 '2014년' 칼럼까지 삭제

```
cctvSeoul = cctvSeoul.drop(['2011년 이전', '2012년', '2013년', '2014년'], axis=1)
cctvSeoul.head()
```

#### '2014년 이전' 칼럼을 '2015년' 칼럼 왼쪽으로 보내기

```
cctvSeoul= cctvSeoul[['소계', '2014년 이전', '2015년', '2016년', '2017년', '2018년']]
cctvSeoul.head()
```

#### CCTV 최근 증가율 계산 ((2015+2016+2017+2018)/2014년 이전) * 100

```
cctvSeoul['최근증가율'] = (cctvSeoul['2015년'] + cctvSeoul['2016년'] + cctvSeoul['2017년'] + cctvSeoul['2018년'])/cctvSeoul['2014년 이전'] * 100
cctvSeoul.head(5)
```

```
# 최근 증가율이 큰 순서대로 내림차순 정렬
cctvTop5 = cctvSeoul.sort_values(by='최근증가율', ascending=False).head(5)
cctvTop5
```

### (2) txt 파일 열기

#### 내 드라이브

```
popuSeoul = pd.read_csv('/gdrive/My Drive/빅데이터교육/seoul_population.txt', sep ='\t')
popuSeoul.head()
```

#### txt 파일 데이터에서 자치구, 합계, 한국인, 등록외국인, 65세이상고령자 열만 추출하기

```
popuSeoul = popuSeoul.iloc[:, [1, 3, 6,  9, 13]]
popuSeoul.head()
```

#### 칼럼 이름 바꾸기

```
popuSeoul.columns= ['자치구', '인구수', '한국인', '외국인', '고령자']
popuSeoul.head(3)
```

#### 합계 행을 맨 아래로 보내기

```
newRowIdx = [x for x in range(1, 26)]
newRowIdx.append(0)
popuSeoul = popuSeoul.reindex(index=newRowIdx)
popuSeoul.head()
```

#### '자치구' 열을 인덱스로 사용

```
# '자치구' 칼럼의 데이터를 리스트로 저장
guList = list(popuSeoul['자치구'])
# '자치구' 칼럼 삭제 후 인덱스를 저장해 두었던 '자치구' 리스트로 사용
del popuSeoul['자치구']
popuSeoul.index = guList
popuSeoul.head()
```

#### '외국인 비율'과 '고령자 비율' 계산

```
popuSeoul['외국인 비율'] = popuSeoul['외국인'] / popuSeoul['인구수'] * 100
popuSeoul['고령자 비율'] = popuSeoul['고령자'] / popuSeoul['인구수'] * 100
popuSeoul.head()
```

#### 천 단위 쉼표(,)를 없애고 자료형을 문자에서 숫자로 변경하기

```
# 천 단위 구분 기호(,) 없애기
delComma = lambda x : x.replace(',', '')
popuSeoul['인구수'] = popuSeoul['인구수'].apply(delComma)
popuSeoul['한국인'] = popuSeoul['한국인'].apply(delComma)
popuSeoul['외국인'] = popuSeoul['외국인'].apply(delComma)
popuSeoul['고령자'] = popuSeoul['고령자'].apply(delComma)
popuSeoul.head(5)
```

```
# 문자열을 수로 변경하고 외국인 비율과 고령자 비율 계산
popuSeoul['인구수'] = pd.to_numeric(popuSeoul['인구수'])
popuSeoul['한국인'] = pd.to_numeric(popuSeoul['한국인'])
popuSeoul['외국인'] = pd.to_numeric(popuSeoul['외국인'])
popuSeoul['고령자'] = pd.to_numeric(popuSeoul['고령자'])

popuSeoul['외국인 비율'] = popuSeoul['외국인'] / popuSeoul['인구수'] * 100
popuSeoul['고령자 비율'] = popuSeoul['고령자'] / popuSeoul['인구수'] * 100

popuSeoul.head()
```

## 인덱스(구) 기준으로 두 데이터 프레임을 합치기

#### CCTV 현황 데이터 프레임에서 '소계'와 '최근증가율' 칼럼만 추출하여 새로운 데이터 프레임 만들기

```
absCCTV = cctvSeoul.loc[:, ['소계', '최근증가율']]
absCCTV.head()
```

#### absCCTV의 칼럼 이름 바꾸기

```
absCCTV.columns = ['CCTV수', '최근증가율']
absCCTV.head()
```

#### absCCTV와 popuSeoul 데이터 프레임 병합하기

```
seoul = pd.merge(absCCTV, popuSeoul, left_index=True, right_index=True, how='left')
seoul.head()
```
