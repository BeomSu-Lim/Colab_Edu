## 난수를 원소로 가지는 배열 생성

np.random.randint(n) : 0 이상 n 미만의 정수형 난수 1개를 생성

```
import numpy as np
# 0에서 9 사이의 랜덤한 정수 1개 생성
randArr1 = np.random.randint(10)
randArr1
```

np.random.randint(m, n) : m 이상 n 미만의 정수형 난수 1개를 생성

```
# 10에서 99 사이의 랜덤한 정수 1개 생성
randArr1 = np.random.randint(10, 100)
randArr1
```

np.random.rand(n) : 0 이상 1 미만의 임의의 실수 n개를 균등한 분포로 선택하여 numpy array 생성<br>
n의 크기가 커질수록 분포는 균등해진다.

---



```
# random.rand(n) 함수 : [0, 1) 사이의 실수형 난수 n개를 균일한 분포로 취하여 배열 생성
randArr1d = np.random.rand(20)
randArr2d = np.random.rand(20).reshape(4, 5)
randArr2d
```

np.random.randn(n) : 표준 정규 분포(평균:0, 표준편차:1, 평균을 중심으로 좌우대칭 종모양)를 따르는 n개의 실수 선택하여 numpy array를 생성

```
randnArr1d = np.random.randn(20)
randnArr2d = np.random.randn(20).reshape(4, 5)
randnArr2d
```

**Numpy와 Matplotlib를 이용한 sin 곡선 그리기**

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

**sin 곡선과 cos 곡선을 함께 그리기**

```
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 200)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.title("Sin & Cos Curve")
plt.plot(X, Y1, 'r--', label='sin curve')
plt.plot(X, Y2, 'b--', label='cos curve')
plt.legend()

plt.show()
```

##3. 막대 그래프 작성하기

막대 그래프는 각 항목 값들의 크기 비교를 쉽게 할 수 있어요.

```
import numpy as np
import matplotlib.pyplot as plt

x = [x for x in range (1, 11)]      # x 리스트의 원소는 10개, 길이 역시 10
y = [np.random.randint(10, 50) for i in range (len(x))]   # 10 <= y < 50, 10개
plt.xticks(x)
plt.bar(x, y)
plt.show()
```

**다중 막대 그래프 그리기**

```
import numpy as np
import matplotlib.pyplot as plt

x = [x for x in range (1, 11)]
y1 = [np.random.randint(10, 50) for i in range (len(x))]
y2 = [np.random.randint(10, 50) for i in range (len(x))]
plt.xticks(x)
plt.bar(x, y1, color = 'yellow')
plt.bar(x, y2, color = 'green')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([x for x in range (1, 11)]).astype(np.float32)
y1 = [np.random.randint(10, 50) for i in range (len(x))]
y2 = [np.random.randint(10, 50) for i in range (len(x))]
plt.xticks(x)
plt.bar(x, y1, color = 'yellow', width=0.4)
plt.bar(x+0.4, y2, color = 'green', width=0.4)  # 0.4 떨어진 곳에서 두번째 막대 생성
plt.show()
```

##4. 원형 그래프 그리기

원형 그래프는 전체에 대하여 각 항목의 값이 차지하는 비율을 확인할 수 있는 차트입니다. 

```
import matplotlib.pyplot as plt

expenditure = [300000, 150000, 100000, 70000, 30000, 250000]
item = ['A', 'B', 'C', 'D', 'E', 'F']

plt.pie(expenditure, labels = item, autopct='%.1f%%')
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt

expenditure = [300000, 150000, 100000, 70000, 30000, 250000]
item = ['A', 'B', 'C', 'D', 'E', 'F']

plt.pie(expenditure, labels = item, startangle=90, autopct='%.1f%%')
plt.legend()
plt.show()
```

```
import matplotlib.pyplot as plt

expenditure = [300000, 150000, 100000, 70000, 30000, 250000]
item = ['A', 'B', 'C', 'D', 'E', 'F']

plt.pie(expenditure, labels = item, startangle=90, shadow = True, explode=(0.1, 0, 0, 0, 0, 0), autopct='%.1f%%')
plt.legend()
plt.show()
```
