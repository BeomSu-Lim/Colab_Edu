## 관계연산자: >, >=, <, <=, ==

```
a, b = 5, 3
a == b
# 관계연산자의 결과는 True 또는 False
```

## 논리연산자: and, or, not

```
10 >= 8 and 6 < 5
# True and False = False
```

```
10 >= 8 or 6 < 5
# True or False = True <- 처리속도 더 빠름
# False of True = True
```

```
not (10 > 5)
```

## (1) 문자열
  - 문자열은 한 개 이상의 연속된 문자를 의미
  - 문자열은 반드시 작은 따옴표(' ')나 큰 따옴표(" ")의 싸으로 둘러싸임

```
letter = 'A'
numChar = "4"
message = 'Good Morning!'
```

### 인덱싱 (Indexing)
인덱싱이란 문자열에 []를 붙여서 문자열 중 하나의 문자를 추출하는 것

```
word = 'python' # 앞에서 맨 처음(0부터) 문자 추출
word[0]
```

```
word[-1] # 뒤에서 첫 번째 문자 추출
```

```
word[0:2] # 0, 1번째 문자 추출
# 문자열의 슬라이싱 (m:n) m <= x < n
# word[:2]와 동일
```

```
word[2:5] # 2, 3, 4번째 문자 추출
# word[2:-1]과 동일: 2번 방부터 맨 뒤칸의 앞 방까지 추출
```

```
word[:2] + word[2:] # 2번방 이전까지 + 2번방부터 끝까지 = 전체
```

# 선택문과 반복문

## (1) 선택문

단순 if문

```
x = int(input('x = ? '))
y = int(input('y = ? '))
if x < y :
    result = 'x가 y보다 더 작습니다.'
    print(result)
```

```
x = int(input('x = ? '))
y = int(input('y = ? '))
if x < y:
    result = 'x가 y보다 더 작습니다.'
else :
    result = 'x가 y보다 더 큽니다.'
print(result)
```

if ~ elif ~ else ~ 문

```
x = int(input('x = ? '))
y = int(input('y = ? '))
if x < y:
    result = 'x가 y보다 더 작습니다.'
elif x > y :
    result = 'x가 y보다 더 큽니다.'
else:
    result = "x와 y는 같습니다."
print(result)
```

**단순 if문 연습**
원의 넓이 계산: 단, 반지름의 길이가 0보다 커야 원이 될 수 있음

```
import math
radius = float(input('반지름의 길이 = '))
if radius > 0:      # 반지름의 길이가 0 초과일때만 원 넓이 출력
    area = radius ** 2 * math.pi
    print("원의 넓이 = %.3f" %area)

# 다른 코드 수행 부분
print('end of if')
```

**if else문 연습**

```
import math
radius = float(input('반지름의 길이 = '))
if radius > 0:      # 반지름의 길이가 0 초과일때만 원 넓이 출력
    area = radius ** 2 * math.pi
    print("원의 넓이 = %.3f" %area)
else:
    print("반지름의 길이는 0보다 커야합니다.")
```

```
# 홀수/짝수 판정
num = int(input('정수를 입력하시오. : '))
if num % 2 == 1:
    result = '홀수'
else:
    result = '짝수'
result
```

**if elif else문 연습**
학생의 성적을 입력받아서 성적에 해당하는 등급 출력

```
score = int(input('점수 입력: '))
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'
print('등급: ', grade)
```
