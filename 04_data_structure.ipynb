<b>리스트 연산

```python
# 최댓값 찾기
score = [36, 78, 55, 49, 68, 64, 80, 93, 88, 80]
maxScore = score[0]
for x in range(1, len(score)):
    if score[x] > maxScore:
        maxScore = score[x]

maxScore
```

```python
# 조건을 만족하는 항목 모두 찾기
score = [36, 78, 55, 49, 68, 64, 80, 93, 88, 84]
excellent = []
for value in score:
    if value >= 80:
        excellent.append(value)
        
excellent
```

**filter() 함수를 이용하여 조건을 만족하는 항목 모두 찾기**

```python
score = [36, 78, 55, 49, 68, 64, 80, 93, 88, 84]
excellent = list(filter(lambda x : x >= 80, score))
excellent
```

<b>내장 순차 자료형 함수

**리스트에 map() 함수 사용 (복습)**<br>
**map() 함수 :** 반복 가능(iterable)한 객체(리스트, 튜플 등)를 입력 받아서, 각 요소에 지정한 함수를 적용할 수 있는 함수

예를 들어, 다음과 같은 리스트 a가 있다고 하자.<br>
a = [1, 2, 3, 4, 5]<br>
리스트 a의 원소를 모두 2배씩 증가시킨 리스트 b를 생성하려면 어떻게 해야할까?

```python
def times2x(x):
    return 2*x

numList = [1, 2, 3, 4, 5]
newList = []
for n in numList:
    newList.append(times2x(n))

newList
```

**map() 함수 이용**

```python
def times2x(x):
    return 2*x

numList = [1, 2, 3, 4, 5]
newList = map(times2x, numList)

list(newList)
```

```python
# 실수 리스트 원소를 한꺼번에 정수로 변환
fl = [3.6, 4.3, 7.8, 9.2, 10.5, 8.7]
il = list(map(int, fl))
il
```

```python
# 정수 리스트의 원소를 한꺼번에 문자열로 변환
intList = [x for x in range(10)]
strList = list(map(str, intList))
strList
```

###3. 사전(dictionary)

####사전
* 키(key)와 값(value)의 쌍으로 구성된 순서가 없는 자료구조
  * 순서가 없기 때문에 인덱스로는 접근할 수 없으며 반드시 키를 이용하여 접근해야 한다.
  * 값은 중복될 수 있지만 키는 중복될 수 없다. 키가 중복되면 마지막 값으로 덮어쓰기가 된다.
* 키를 이용하여 값을 신속하게 찾아내는 해시테이블(Hash Table) 구조


####사전 선언하기
중괄호({})를 이용하여 사전을 선언한다. 

```python
# 사전 구조의 학생 성적 입력 (키 : 번호)
score1 = {1:88, 2:96, 3:73, 4:68, 5:72, 6:80}
score1
```

```python
# 사전 구조의 학생 성적 입력 (키 : 이름)
score2 = {'강은미':88, '김영민':96, '도민준':73, '박현우':68, '이기쁨':72, '하지은':80}
score2
```

```python
# dict() 함수를 이용한 사전 선언
score3 = dict(강은미= 88, 김영민= 96, 도민준= 73, 박현우= 68, 이기쁨= 72, 하지은= 80)
score3
```

```python
circle = {'center' : (10, 10), 'radius' : 5}
circle
```

####사전의 원소에 접근
키를 이용하여 접근한다. 순서가 없기 때문에 인덱스로는 접근할 수 없다.

```python
# 박현우의 성적을 알고 싶은데 score2 사전의 3번 인덱스로 접근하면 될까?
score2[3]
```

```python
score2['박현우']
```

```python
# 그런데 이건 왜 될까?
score1[3]
```

####사전 원소 값(value)의 변경

```python
score1[3] = 75
score2['이기쁨'] = 78
print(score1)
print(score2)
```

**update() 함수를 이용한 원소의 변경**

```python
#키가 문자일 때 update(키=값)
score2.update(도민준=90) 
score2
```

```python
# 사전의 여러 원소를 한꺼번에 변경
score2.update(강은미=89, 하지은=83)
score2
```

```python
#키가 숫자일 때 update(키 : 값)
score1.update({2 : 100}) 
score1
```
