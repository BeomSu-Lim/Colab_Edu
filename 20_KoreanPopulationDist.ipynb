```
import numpy as np
import pandas as pd
import json
import folium
```

```
popuData = pd.read_excel('/gdrive/My Drive/빅데이터교육/202010_202010_주민등록인구및세대현황_시도.xlsx')
popuData
```

```
popuData.drop([0, 1, 2], inplace=True)
popuData
```

```
popuData.drop(['※ 매월 말일자 통계 현황'], axis=1, inplace=True)
popuData
```

```
popuData.columns = ['행정구역', '총인구수', '세대수', '세대당 인구수', '남자 인구수', '여자 인구수', '남여 비율']
popuData
```

##지역명의 공백 제거(중요, 중요, 중요, 중요!!!!!!!)

```
delSpace = lambda x : x.replace(' ', '')
popuData.iloc[:, 0] = popuData.iloc[:, 0].apply(delSpace)
```

##천 단위 구분 기호 제거

```
# 천 단위 구분 기호(,) 없애기
delComma = lambda x : x.replace(',', '')
for i in range(1, 6):
    popuData.iloc[:, i] = popuData.iloc[:, i].apply(delComma)
popuData
```

##문자열을 수로 변경

```
# 문자열을 수로 변경
for i in range(1, 7):
    popuData.iloc[:, i] = pd.to_numeric(popuData.iloc[:, i])
```

```
print(popuData.dtypes) #각 칼럼의 데이터타입 확인
```

```
local = list(popuData['행정구역'])
popuData.drop(['행정구역'], axis=1, inplace=True)
popuData.index = local
popuData
```

```
with open('/gdrive/My Drive/빅데이터교육/TL_SCCO_CTPRVN.json', encoding='utf-8') as json_file:
    korea_geo = json.load(json_file)
korea_geo
```

```
popuKorea = folium.Map(location=[36, 127], tiles = 'cartodbpositron', zoom_start=7)
folium.Choropleth(
    geo_data=korea_geo,
    name='Population Distribution by Region in Korea',
    data= popuData['총인구수'], 
    columns=[popuData.index, popuData['총인구수']],
    key_on='feature.properties.CTP_KOR_NM',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacitruety=0.3,
    highlight=True,
    legend_name = '지역별 총인구',
).add_to(popuKorea)
popuKorea
```
